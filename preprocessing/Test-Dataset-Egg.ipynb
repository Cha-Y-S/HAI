```python
import pandas as pd
df = pd.read_csv('egg.csv')
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df1 = df.T
```

```python
df1
```

```python
df1.columns
```

```python
# 첫번째 사진의 output에 대해 dataframe을 생성
tmp = df1[0].str.split(",")
df2 = pd.DataFrame(data=[tmp.str.get(j) for j in range(92)], index=range(92))
df2.loc[92] = "Egg"
df2 = df2.T
df2
```

```python
# 2번째 사진부터 마지막 사진까지 각각 output을 구하고 그에 대한 dataframe을 생성하면서
# 동시에 df2(첫번째 사진의 dataframe)에 이어 붙인다.
for i in range(1, len(df1.columns)):
    temp = df1[i].str.split(",")
    df3 = pd.DataFrame(data=[temp.str.get(j) for j in range(92)], index=range(92))
    df3.loc[92] = "Egg"
    df3 = df3.T
    df2 = pd.concat([df2,df3])
```

```python
df2 = df2.reset_index(drop=True)
```

```python
df2
```

```python
# 최종 csv파일을 얼굴형에 맞는 이름으로 저장한다.
# 계란형 - Egg, 둥근형 - Circle, 각진형 - Angled, ...
df2.to_csv('test-egg.csv',index=False)
```
